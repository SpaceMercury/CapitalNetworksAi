```python
import PySimpleGUI as sg
import random as rd

def advisor(stock, features):
    return f"Advice for {stock}: Based on the features {', '.join(features.keys())}, here's some advice..."

def Chat(txt):
    return f"Chatbot response to '{txt}': This is a mock response."

def get_best_5_stocks():
    return ['AAPL', 'MSFT', 'GOOG', 'FB', 'AMZN']

def get_the_best_stock_features_that_user_likes():
    return ['Price', 'Volume', 'Market Cap', 'P/E Ratio', 'Dividend Yield']

def get_stock_info(stock):
    return {feature: round(rd.uniform(10, 1000), 2) for feature in get_the_best_stock_features_that_user_likes()}

stocks_features = {stock: get_stock_info(stock) for stock in get_best_5_stocks()}

# Define colors and font styles
bg_color = '#f0f0f0'
text_color = '#003f5c'
button_color = ('white', '#007f5f')
font_large = ("Helvetica", 20)
font_medium = ("Helvetica", 16)
font_small = ("Helvetica", 14)

window_width, window_height = 1424, 968
layout = [
    [sg.Text("For You", size=(int(window_width/12), 1), font=font_large, text_color=text_color)],
    [sg.HorizontalSeparator()],
    [sg.Text("", font=font_medium)],
    [sg.Button(stock, size=(20, 2), font=font_small, button_color=button_color) for stock in get_best_5_stocks()],
    [sg.Text("Info:", font=font_medium)],
    [sg.Text("", font=font_small, key='-FEATURES-', size=(50, 5))],
    [sg.Text("", key='-ADVISOR-', size=(int(window_width/12), 10), font=font_small)],
    [sg.HorizontalSeparator()],
    [sg.Text("Ask Shin:", font=font_medium)],
    [sg.InputText(key='-CHATINPUT-', size=(int(window_width/15), 1), font=font_small), sg.Button('Chat', size=(10, 1), font=font_small, button_color=button_color)],
    [sg.Text("", key='-CHATRESPONSE-', size=(int(window_width/12), 3), font=font_small)],
    [sg.Button('Exit', size=(4, 2), font=font_small, pad=((5,0), (0,0)), button_color=('white', 'red'))]
]

window = sg.Window('Powered by Ai’s Networks Group', layout, size=(window_width, window_height), background_color=bg_color)

while True:
    event, values = window.read()
    if event in (None, 'Exit'):
        break

    if event in get_best_5_stocks():
        stock_features = stocks_features[event]
        feature_text = "\n".join([f"{k}: {v}" for k, v in stock_features.items()])
        window['-FEATURES-'].update(feature_text)
        advice = advisor(event, stock_features)
        window['-ADVISOR-'].update(advice)

    if event == 'Chat':
        chat_response = Chat(values['-CHATINPUT-'])
        window['-CHATRESPONSE-'].update(chat_response)

window.close()
```
